## 2. Roots y formatives

```python
## roots!

df_map_roots = {}

for name in df_map.keys():
    df = df_map[name]
    columns_root = [n for n in df.columns if (n == 'Unnamed: 1' or n.startswith('ROOT'))]
    
    DF = df[columns_root][1:]
    l = list(DF.agg(list, axis=1).to_dict().values())
    D = {}
    for item in l:
        D[item[0]]=item[1:]
    df_map_roots[name]=D
```

```python
## formatives!

df_map_for = {}

for name in df_map.keys():
    df = df_map[name]
    columns_for = [n for n in df.columns if (n == 'Unnamed: 1' or n.startswith('FORMATIVE'))]
    
    DF = df[columns_for][1:]
    l = list(DF.agg(list, axis=1).to_dict().values())
    D = {}
    for item in l:
        D[item[0]]=item[1:]
    df_map_for[name]=D
```

```python
from math import isnan

roots = {language:{} for language in df_map_roots['head'].keys()}

for name in df_map_roots.keys():
    D_name = df_map_roots[name]
    D_name = {k: D_name[k] for k in D_name if not pd.isna(k)}
    for language in D_name.keys():
        roots[language][name]=[int(x) for x in D_name[language]]
```

```python
for name in roots.keys():
    print(name,len(roots[name]))
```

```python
nexus_roots = {}

for name in roots.keys():
    nexus = []
    for key in roots['Shipibo_Konibo'].keys():
        if key in roots[name].keys():
            nexus+=roots[name][key]
        else:
            nexus+=['?']*len(roots['Shipibo_Konibo'][key])
    nexus_roots[name]=''.join([str(x) for x in nexus])
```

```python
import numpy as np

## distancia de Hamming 

def hamming(lengua1, lengua2, root=True):
    ## distances
    H = []
    
    ## rasgos
    if root == True:
        features_lengua1=roots[lengua1]
        features_lengua2=roots[lengua2]
    else:
        features_lengua1=form[lengua1]
        features_lengua2=form[lengua2]
    
    ## hamming!
    for feature in features_lengua1.keys():
        d=0
        n=0
        if feature in features_lengua2.keys() and len(features_lengua1[feature])>0: 
            for i in range(len(features_lengua1[feature])):
                if features_lengua1[feature][i] != features_lengua2[feature][i]:
                    d += 1.0
                n += 1.0
            H+=[d/n]
            
    return np.mean(H)
```

```python
lenguas_root = {lengua:{} for lengua in roots.keys()}

for lengua in lenguas_root.keys():
    for lengua_lengua in lenguas_root.keys():
        lenguas_root[lengua][lengua_lengua]=hamming(lengua,lengua_lengua)
```

```python
## en X guardamos todas las listas de rasgos

X=[]
for language in lenguas_root.keys():
    X+=[list(lenguas_root[language].values())]
```

```python
#from sklearn.manifold import TSNE
#model = TSNE(n_components=2, perplexity=5, random_state=0, metric='precomputed',square_distances=True)
#X = model.fit_transform(X)

from sklearn.decomposition import PCA
transformer = PCA(n_components=2)
X = transformer.fit_transform(X)
```

```python
print(transformer.explained_variance_ratio_)
```

```python
x=list(zip(*list(X)))[0]
y=list(zip(*list(X)))[1]
```

```python
import matplotlib.pyplot as plt

cmap = plt.get_cmap('RdBu')

fig, ax = plt.subplots(dpi=800)

for i in range(len(x)):
    ax.annotate(list(lenguas_root.keys())[i], color='k', xy=(x[i],y[i]), fontsize=3.5, alpha=0.65)
#    ax.annotate(labels[i], weight='demi', color='r', xy=(x[i]-1.5,y[i]-1.5), fontsize=4, alpha=0.85)

ax.plot([x[i] for i in [list(lenguas_root.keys()).index(language) for language in lenguas_root.keys()]],[y[i] for i in [list(lenguas_root.keys()).index(language) for language in lenguas_root.keys()]],'o',color='orchid',markersize=5,markeredgewidth=0.5,markeredgecolor='k',alpha=0.75,fillstyle='full',clip_on=True)

plt.grid(False)
plt.title('Roots',fontsize=10)
ax.set_yticks([])
ax.set_xticks([])
#plt.legend(loc='best',fontsize=7)
plt.ylabel(r'PCA 2',fontsize=10)
plt.xlabel(r'PCA 1',fontsize=10)
plt.rcParams.update({'font.size': 10})
plt.savefig('root_2D.pdf', format='pdf', transparent=True, bbox_inches='tight',dpi=1024)
plt.show()
```
